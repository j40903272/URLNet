```python
import numpy as np
import pandas as pd
```

```python
black = []
with open("../taipei/black.csv", "r", encoding="utf-8") as f:
    for line in f:
        black.append(line.strip().split(','))
```

```python
white = []
with open("../taipei/white.csv", "r", encoding="utf-8") as f:
    for line in f:
        white.append(line.strip().split(','))
```

```python
total = min(len(black), len(white))
validation = 0.001
split = int(total*(1-validation))

np.random.seed(87)
# np.random.shuffle(black)
# np.random.shuffle(white)


train = []
label = []
for i in black[1:split]:
    train.append(i[0].strip())
    label.append(0) # malicious
for i in white[1:split]:
    train.append(i[0].strip())
    label.append(1) # benign
train_df = pd.DataFrame({"url":train, "label":label})


val = []
label = []
for i in black[split:total]:
    val.append(i[0].strip())
    label.append(0) # malicious
for i in white[split:total]:
    val.append(i[0].strip())
    label.append(1) # benign
val_df = pd.DataFrame({"url":val, "label":label})


test = []
label = []
for i in black[total:]:
    test.append(i[0].strip())
    label.append(0) # malicious
for i in white[total:]:
    test.append(i[0].strip())
    label.append(1) # benign
test_df = pd.DataFrame({"url":test, "label":label})

print(len(train_df), len(val_df), len(test_df))
```

```python
train_df.to_csv("train_10000.txt", index=False, header=None)
val_df.to_csv("val_10000.txt", index=False, header=None)
test_df.to_csv("test_10000.txt", index=False, header=None)
```
